```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
```

## Load and plot data

```python
df = pd.read_csv('data\data.csv', encoding='utf-8', sep='\t'); df[0:10]
```

```python
plt.figure(figsize=(7,5))
plt.scatter(x=df.Height, 
            y=df.Foot, 
            color='black',
            alpha=.3)
plt.xlabel('Height')
plt.ylabel('Foot size')
plt.show()
```

## Standarization & Min-Max normalization

**Standarization**

$$
z = \frac{x - \mu}{\sigma}
$$

where

$$
\mu = \frac{1}{N}\sum_{i=1}^{N}(x_{i})
$$

and

$$
\sigma = \sqrt{\frac{1}{N}\sum_{i=1}^{N}(x_{i}-\mu)^{2}}
$$

**Min-Max normalization**

$$
X_{norm} = \frac{X-X_{min}}{X_{max}-X_{min}} 
$$

where

$$
X = \{x_{1}, x_{2}... x_{n}\}
$$

```python
X = np.asarray(df.Height)
Y = np.asarray(df.Foot)

z_scores_X = (X - X.mean()) / X.std()
z_scores_Y = (Y - Y.mean()) / Y.std()

minmax_X = (X - X.min()) / (X.max() - X.min())
minmax_Y = (Y - Y.min()) / (Y.max() - Y.min())
```

## Visualizations

```python
plt.figure(figsize=(14,10))
plt.scatter(x=X, 
            y=Y,
            color='black', 
            label='Original data', 
            alpha=0.3)

plt.scatter(x=z_scores_X, 
            y=z_scores_Y, 
            color='red',
            label='Standardized data $(\mu=0, \; \sigma=1)$', 
            alpha=0.3)

plt.scatter(x=minmax_X, 
            y=minmax_Y,
            color='blue', 
            label='Min-max scaled data', 
            alpha=0.3)

plt.title('Height vs. foot size')
plt.xlabel('Height')
plt.ylabel('Foot size')
plt.legend(loc='upper left')    
plt.show()
```

```python
plt.figure(figsize=(14,10))

plt.subplot(221)
plt.scatter(x=X, 
            y=Y,
            color='black', 
            label='Original data', 
            alpha=.3)
plt.legend(loc='upper left')

plt.subplot(222)
plt.text(.5, .5,
         'Effect of standardization',
         horizontalalignment='center',
         verticalalignment='center',
         fontsize=20,
         alpha=.7)
plt.xticks(())
plt.yticks(())

plt.subplot(223)
plt.scatter(x=z_scores_X, 
            y=z_scores_Y, 
            color='red',
            label='Standardized data $(\mu=0, \; \sigma=1)$', 
            alpha=.3)
plt.legend(loc='upper left') 

plt.subplot(224)
plt.scatter(x=minmax_X, 
            y=minmax_Y,
            color='blue', 
            label='Min-max scaled data', 
            alpha=.3)
plt.legend(loc='upper left') 

plt.show()
```
